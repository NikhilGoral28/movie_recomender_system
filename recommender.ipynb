```python
#keeping only the required columns
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(2)
```

```python
# data preprocessing

import ast
```

```python
#Extract Relevant Features
def convert(obj):
    L = []
    # Only apply ast.literal_eval if obj is a string
    if isinstance(obj, str):
        items = ast.literal_eval(obj)
    else:
        items = obj
    for i in items:
        L.append(i['name'] if isinstance(i, dict) and 'name' in i else i)
    return L

# applying the convert function to genres and keywords only if they are not already lists
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)

movies['genres']
```

```python
movies['keywords']
```

```python
# Top 3 cast members only
def get_top_cast(obj):
    L = []
    count = 0
    if isinstance(obj, str):
        items = ast.literal_eval(obj)
    else:
        items = obj
    for item in items:
        if count < 3:
            if isinstance(item, dict) and 'name' in item:
                L.append(item['name'])
                count += 1
            elif isinstance(item, str):
                # If already a string (name), just append
                L.append(item)
                count += 1
        else:
            break
    return L

movies['cast'] = movies['cast'].apply(get_top_cast)

movies['cast']
```

```python
# process the reviews
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else [])
```

```python
movies['overview']
```

```python
#combine all the features into a single string

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'].apply(lambda x: [x])
```

```python
movies['tags']
```

```python
#converting to string
movies['tags'] = movies['tags'].apply(lambda x: ' '.join(map(str, x)))
```

```python
movies['tags']
```

```python
#final dataset
final_movies = movies[['movie_id', 'title', 'tags']]
final_movies['tags'] = final_movies['tags'].apply(lambda x: x.lower())
```
